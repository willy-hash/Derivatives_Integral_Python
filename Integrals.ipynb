1. $$\int_1^4 (x^2 - 9x)dx $$

```python
from sympy import Integral, integrate, symbols

x = symbols('x')

fx = x**2 - 9*x
dx = Integral(fx, x).doit()

Integral(fx, (x, 1, 4)).doit()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate

# Define la variable simbólica y la función
x = symbols('x')
fx = x**2 - 9*x

# Calcula la integral definida
resultado = integrate(fx, (x, 1, 4))
print(f"El valor de la integral definida es: {resultado}")

# Gráfica de la función
# Rango de valores
x_vals = np.linspace(0, 5, 500)  # Rango extendido para contexto
y_vals = x_vals**2 - 9*x_vals

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = x^2 - 9x$", color="blue")
plt.fill_between(x_vals, y_vals, where=(x_vals >= 1) & (x_vals <= 4), color="lightblue", alpha=0.5,
                 label=r"$\int_1^4 f(x) dx$")

# Agregar detalles
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(1, color='red', linestyle='--', label="x=1")
plt.axvline(4, color='red', linestyle='--', label="x=4")
plt.title("Integral definida de $x^2 - 9x$ entre 1 y 4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

8. $$\int_\frac{\pi}{2}^\frac{\pi}{2} (2x - sen2x)dx $$



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, sin, pi

# Define la variable simbólica y la función
x = symbols('x')
fx = 2*x - sin(2*x)

# Calcula la integral definida
resultado = integrate(fx, (x, pi/4, pi/2))
print(f"El valor de la integral definida es: {resultado}")

# Gráfica de la función
# Rango de valores
x_vals = np.linspace(0, np.pi, 500)  # Rango extendido para visualizar la función
y_vals = 2*x_vals - np.sin(2*x_vals)

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = 2x - \sin(2x)$", color="blue")
plt.fill_between(x_vals, y_vals, where=(x_vals >= np.pi/4) & (x_vals <= np.pi/2), color="lightblue", alpha=0.5,
                 label=r"$\int_{\frac{\pi}{4}}^{\frac{\pi}{2}} f(x) dx$")

# Agregar detalles
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(np.pi/4, color='red', linestyle='--', label=r"$x = \frac{\pi}{4}$")
plt.axvline(np.pi/2, color='red', linestyle='--', label=r"$x = \frac{\pi}{2}$")
plt.title(r"Integral definida de $2x - \sin(2x)$ entre $\frac{\pi}{4}$ y $\frac{\pi}{2}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()

```

9. $$\int_0^\frac{\pi}{6}(1 + cos3x)dx  $$

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, cos, pi

# Define la variable simbólica y la función
x = symbols('x')
fx = 1 + cos(3*x)

# Calcula la integral definida
resultado = integrate(fx, (x, 0, pi/6))
print(f"El valor de la integral definida es: {resultado}")

# Gráfica de la función
# Rango de valores
x_vals = np.linspace(0, np.pi/6, 500)  # Cambiado a np.pi
y_vals = 1 + np.cos(3*x_vals)

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = 1 + \cos(3x)$", color="blue")
plt.fill_between(x_vals, y_vals, where=(x_vals >= 0) & (x_vals <= np.pi/6), color="lightblue", alpha=0.5,
                 label=r"$\int_0^{\frac{\pi}{6}} f(x) dx$")

# Agregar detalles
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='red', linestyle='--', label=r"$x = 0$")
plt.axvline(np.pi/6, color='red', linestyle='--', label=r"$x = \frac{\pi}{6}$")
plt.title(r"Integral definida de $1 + \cos(3x)$ entre $0$ y $\frac{\pi}{6}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()

```
