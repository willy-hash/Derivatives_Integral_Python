1.  5x -7

```python
import sympy as sp

# Definimos la variable simbólica
x = sp.symbols('x')

# Definimos la función
f = 5*x -7

# Calculamos la derivada
dfdx = sp.diff(f, x)

# Imprimimos el resultado
print(dfdx)
```

Grafica

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, integrate, lambdify

# Definir la variable simbólica
x = symbols('x')

# Definir la función original
fx = 5*x - 7

# Calcular la derivada y la integral indefinida de la función
derivada_fx = diff(fx, x)

# Convertir las expresiones simbólicas a funciones que se puedan evaluar numéricamente
f_numeric = lambdify(x, fx, 'numpy')
derivada_numeric = lambdify(x, derivada_fx, 'numpy')

# Crear un rango de valores de x para graficar
x_vals = np.linspace(-8, 8, 20)

# Evaluar la función, su derivada y su integral en el rango de x
y_vals = f_numeric(x_vals)
# The following line is changed to fix the dimension mismatch
dy_vals = derivada_numeric(x_vals) * np.ones_like(x_vals)
# We multiply the derivative (which is a constant) by an array of ones
# with the same shape as x_vals.
# This ensures dy_vals has the same dimension as x_vals


# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = 5x - 7$") # Corrected label
plt.plot(x_vals, dy_vals, label=r"$f'(x)$ (Derivada de $f(x)$)", linestyle="--")

# Añadir título y etiquetas
plt.title("Gráfica de $f(x)$, su derivada")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

3. $f(x) = \sqrt{x^4 - 9}$

```python
import sympy as sp

# Definimos la variable simbólica
x = sp.symbols('x')

# Definimos la función
f = sp.sqrt(x**4 - 9)

# Calculamos la derivada
dfdx = sp.diff(f, x)

# Imprimir el resultado usando LATEX
sp.init_printing()

display(dfdx)
```

Grafica

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sqrt, diff, lambdify

# Definir la variable simbólica
x = symbols('x')

# Definir la función original
fx = sqrt(x**4 - 9)

# Calcular la derivada
derivada_fx = diff(fx, x)

# Convertir las expresiones simbólicas a funciones evaluables numéricamente
f_numeric = lambdify(x, fx, 'numpy')
derivada_numeric = lambdify(x, derivada_fx, 'numpy')

# Crear un rango de valores de x para graficar
x_vals = np.linspace(-8, 8, 1000)

# Filtrar valores donde x^4 - 9 >= 0 (para evitar errores con la raíz)
x_vals = x_vals[np.abs(x_vals) >= 3**(1/4)]

# Evaluar la función y su derivada en el rango de x
y_vals = f_numeric(x_vals)
dy_vals = derivada_numeric(x_vals)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = \sqrt{x^4 - 9}$")
plt.plot(x_vals, dy_vals, label=r"$f'(x)$ (Derivada de $f(x)$)", linestyle="--")

# Añadir título, etiquetas y leyenda
plt.title("Gráfica de $f(x)$ y su derivada")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()


```

Sea f(x) = $\frac{x+5}{x+1}$

a)  Evaluar f'(1)

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, lambdify

# Definir la variable simbólica
x = symbols('x')

# Definir la función original
fx = (x + 5) / (x + 1)

# Calcular la derivada de la función
derivada_fx = diff(fx, x)

# Convertir las expresiones simbólicas a funciones evaluables numéricamente
f_numeric = lambdify(x, fx, 'numpy')
derivada_numeric = lambdify(x, derivada_fx, 'numpy')

# Evaluar la derivada en x = 1
valor_derivada_en_1 = derivada_numeric(1)
print(f"El valor de f'(1) es: {valor_derivada_en_1}")

# Crear un rango de valores de x para graficar
x_vals = np.linspace(-8, 8, 500)
x_vals = x_vals[x_vals != -1]  # Excluir el punto x = -1 para evitar división por cero

# Evaluar la función en el rango de x
y_vals = f_numeric(x_vals)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = \frac{x+5}{x+1}$", color="blue")

# Añadir título, etiquetas y leyenda
plt.title("Gráfica de $f(x)$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axvline(x=-1, color="red", linestyle="--", label="Asintota en $x = -1$")
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
